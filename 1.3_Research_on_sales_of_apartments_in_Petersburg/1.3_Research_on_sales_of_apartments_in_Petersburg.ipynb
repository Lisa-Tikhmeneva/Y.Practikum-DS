#### Высота потолков

```python
df['ceiling_height'].value_counts()
```

```python
sorted(df['ceiling_height'].unique())
```

Высота потолков от 1 метра до 100... И есть высота потолков с сантиметровой точностью.

1. Потолки высотой меньше 2,4 метра  пока оставлю без изменений. 2,4 метра это минимальная высота потолков в квартире.
2. Потолки больше 24 метров не существуют, это может быть ошибкой в порядке значения, возникшей при заполнении формы.
3. В потолки больше 6ти метров не верится.
4. Сантиметровая точность вывода потолков излишня, приведём всё к десятисантиметровй точности.

Исправим этот столбец:
Создадим и применим функцию, которая все значения потолков более 24х метров включительно разделит на 10 и округлит все значения до десяти сантиметров.

```python
df['ceiling_height'] = df['ceiling_height'].apply(lambda x: round(x / 10, 1) if x >= 24 else round(x, 1))
```

```python
df['ceiling_height'].value_counts()
```

```python
df['ceiling_height'].isna().sum()
```

Количество различных значений в данном столбце сократилось с 183х до 46. Если посмотреть количество значений, то можно заметить, что в единичных значениях присутствует высота потолков 1.2 метра, 22.6... Скорее всего, это опечатки тех, кто заполнял данные о своей квартире.

Также в столбце есть 9195 пропущенных значений. Возможно можно часть значений восстановить по характерным домам, т.к. известно, в какой серии домов какая высота потолков. Но у нас нет столбца с серией дома, а также жильцы могли уменьшить высоту потолков(установив натяжной потолок, на пример).

Поэтому не вижу возможности восстановить высоту потолка.

#### Жилая площадь в квадратных метрах:

```python
df['living_area'].value_counts()
```

1903 пропущенных значения. Заполним пустые значения медианными значениями площади по параметру "количество комнат"

```python
#Создадим столбец, в котором указаны медианные значения площади для каждой строки согласно количеству комнат в квартире:
median_living_area = df.groupby('rooms')['living_area'].transform('median')
median_living_area
```

```python
# заполним все пустые значения жилой площади на медианные значения из нового столбца
df['living_area'] = df['living_area'].fillna(median_living_area)
```

Площадь помещений с точностью до квадратных сантиметров очень плохо воспринимается, к тому же обычно замеры квартир на проводят с такой точностью. Округлим площадь комнат до десятичной части:

```python
def round_area(area):
    return round(area,1)
```

```python
df['living_area'] = df['living_area'].apply(round_area)
```

```python
df['living_area'].isna().sum()
```

```python
df['living_area'].min()
```

```python
df['living_area'].max()
```

###  день недели, месяц и год публикации объявления

Добавим столбец `weekday_month_year`, в котором будут указаны день_недели-месяц-год публикации объявления

```python
df['datetime'] = pd.to_datetime(df['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')

df['weekday'] = df['datetime'].dt.weekday
df['month']   = df['datetime'].dt.month
df['year']    = df['datetime'].dt.year
```

### этаж квартиры

Добавим столбец `floor_category`, в котором будет указана категория этажа квартиры, варианты — первый, последний, другой

```python
def floor_category(row):
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return ' другой'
```

```python
df['floor_category'] =  df.apply(floor_category, axis=1)
```
